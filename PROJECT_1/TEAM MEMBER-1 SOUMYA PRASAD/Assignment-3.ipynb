## Outliers And Replace The Outliers

### Outlier Detection

```python
sns.boxplot(df.Rings)
```

### Outlier Replacement Using Median

```python
df.median()
```

```python
Q3 = df.Rings.quantile(0.75)
Q1 = df.Rings.quantile(0.25)
IQR = Q3-Q1
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
```

```python
upper_limit
```

```python
lower_limit
```

```python
df = df[df.Rings<upper_limit]
sns.boxplot(df.Rings)

```

```python
df.shape
```

## Check for categorical columns and perform encoding

```python
df.head()
```

### Encoding

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.Sex = le.fit_transform(df.Sex)
df.head()
```

```python
df.describe()
```

### On Hot Encoding

```python
df_main=pd.get_dummies(df,columns=['Sex'])
df_main.head()
```

```python
df_main.describe()
```

### Correlation

```python
df_main.head()
```

## Split the data into dependent and independent variables

### independent variables

```python
X=df_main.drop(columns=['Height'],axis=1)
X.head()
```

### y-target dependent variable

```python
y=df_main.Height
y
```

```python
X.head()
```

### Scaling

```python
from sklearn.preprocessing import scale
```

```python
X_scaled=pd.DataFrame(scale(X),columns=X.columns)
X_scaled.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X_scaled,y, test_size=0.3,random_state=0)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test
```

```python
X_test.shape
```

```python
y_test.shape
```

## Build the Model

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
ridge = Ridge()
lasso = Lasso()
```

```python
ridge.fit(X_train,y_train)
```

```python
lasso.fit(X_train,y_train)
```

### Training and testing the model

```python
y_train_pred = ridge.predict(X_train)
y_train_pred
```

```python
y_test_pred = ridge.predict(X_test)
y_test_pred
```

```python
y_train_predl = lasso.predict(X_train)
y_train_predl
```

```python
y_test_predl = lasso.predict(X_test)
y_test_predl
```

### Evaluation metrics for regression

```python
from sklearn import metrics
```

```python
# MSE

# ridge mse
metrics.mean_squared_error(y_test,y_test_pred)
```

```python
# lasso mse
metrics.mean_squared_error(y_test,y_test_predl)
```

```python
# RMSE

#ridge rmse
np.sqrt(metrics.mean_squared_error(y_test,y_test_pred))
```

```python
# lasso rmse 
np.sqrt(metrics.mean_squared_error(y_test,y_test_predl))
```

```python
# ridge training acc.

metrics.r2_score(y_train,y_train_pred)
```

```python
# testing acc. of ridge

metrics.r2_score(y_test,y_test_pred) 
```

```python
# lasso training  acc

metrics.r2_score(y_train,y_train_predl)
```

```python
# testing acc. of lasso

metrics.r2_score(y_test,y_test_predl) 
```
