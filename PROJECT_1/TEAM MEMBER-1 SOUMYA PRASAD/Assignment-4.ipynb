## Outliers And Replace The Outliers

## Outlier Detection

```python
sns.boxplot(df.iloc[:,3])
```

## Outlier removal-IQR method

```python
Q1 = df.iloc[:,3].quantile(0.25)
Q3 = df.iloc[:,3].quantile(0.75)
IQR = Q3-Q1
```

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
upper_limit
```

```python
lower_limit
```

```python
df = df[df.iloc[:,3]<upper_limit]
```

```python
sns.boxplot(df.iloc[:,3])
```

```python
df.shape
```

## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df)
data_scaled[0:5]
```

## Perform Clustering algorithm

```python
from sklearn import cluster
```

```python
from sklearn.cluster import KMeans
km = KMeans()
res = km.fit_predict(data_scaled)
res
```

```python
data = pd.DataFrame(data_scaled, columns = df.columns)
data.head()
```

```python
data['kclus']  = pd.Series(res)
data.head()
```

## Independent and dependent variable splits

## Independent variables

```python
X= data.iloc[:,:-1]
X.head()
```

## Dependent variables


```python
y = data.kclus
y
```

```python
X.head()
```

## Scaling the independent variables

```python
from sklearn.preprocessing import scale
X_scaled = pd.DataFrame(scale(X),columns=X.columns)
X_scaled.head()
```

## Spliting the data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X_scaled,y, test_size=0.3,random_state=0)
```

## Building the model

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
ridge = Ridge()
lasso = Lasso()
```

```python
ridge.fit(X_train,y_train)
```

```python
lasso.fit(X_train,y_train)
```

## Training and Testing the model

```python
y_train_pred = ridge.predict(X_train)
y_train_pred
```

```python
y_test_pred = ridge.predict(X_test)
y_test_pred
```

```python
y_train_predl = lasso.predict(X_train)
y_train_predl
```

```python
y_test_predl = lasso.predict(X_test)
y_test_predl
```

## Evaluation Metrics for Regression

```python
from sklearn import metrics
```

```python
# MSE

# ridge mse
metrics.mean_squared_error(y_test,y_test_pred)
```

```python
# lasso mse
metrics.mean_squared_error(y_test,y_test_predl)
```

```python
# RMSE

#ridge rmse
np.sqrt(metrics.mean_squared_error(y_test,y_test_pred))
```

```python
# lasso rmse 
np.sqrt(metrics.mean_squared_error(y_test,y_test_predl))
```

```python
# ridge training acc.

metrics.r2_score(y_train,y_train_pred)
```

```python
# testing acc. of ridge

metrics.r2_score(y_test,y_test_pred) 
```

```python
# lasso training  acc

metrics.r2_score(y_train,y_train_predl)
```

```python
# testing acc. of lasso

metrics.r2_score(y_test,y_test_predl) 
```
